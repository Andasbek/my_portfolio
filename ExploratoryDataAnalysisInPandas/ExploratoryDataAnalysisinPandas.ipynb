Содержание

В этом наборе данных у нас есть исторические данные о населении для каждой страны/территории в мире по различным параметрам, таким как площадь страны/территории, название континента, название столицы, плотность населения, темп роста населения, ранжирование по населению, процент мирового населения и т. д.

```python
# Импорт необходимых библиотек
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Чтение данных из CSV-файла
df = pd.read_csv("world_population.csv")
```

```python
# Вывод первых нескольких строк DataFrame
df.head()
```

```python
# Установка опций отображения чисел с плавающей запятой
pd.set_option("display.float_format", lambda x: "%.2f" % x)
```

```python
# Вывод информации о данных
df.info()
```

```python
# Вывод описательных статистик
df.describe()
```

```python
# Подсчет количества пропущенных значений
df.isna().sum()
```

```python
# Подсчет количества уникальных значений
df.nunique()
```

```python
# Сортировка DataFrame по столбцу "2022 Population" в порядке убывания и вывод первых 10 строк
df.sort_values(by="2022 Population", ascending=False).head(10)
```

```python
# Вывод названий столбцов DataFrame
df.columns
```

```python
# Фильтрация числовых столбцов и построение матрицы корреляции
numeric_columns = df.select_dtypes(include=[np.number])
correlation_matrix = numeric_columns.corr
correlation_matrix()
```

```python
# Построение тепловой карты матрицы корреляции
numeric_columns = df.select_dtypes(include=['float'])
sns.heatmap(numeric_columns.corr(), annot=True)
plt.rcParams['figure.figsize'] = (24, 7)
plt.show()
```

```python
df
```

```python
# Группировка данных по континентам и вычисление средних значений числовых столбцов
df.groupby("Continent").mean(numeric_only=True).sort_values(by="2022 Population", ascending=False)
```

```python
# Создание DataFrame с средними значениями числовых столбцов по континентам сгруппированных по годам и построение графика
df2 = df.groupby('Continent')[['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']].mean().sort_values(by="2022 Population",ascending=False)
df2
```

```python
df2.plot();
```

```python
# Транспонирование DataFrame и построение графика
df3 = df2.transpose()
df3
```

```python
df3.plot()
```
